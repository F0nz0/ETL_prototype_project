# Analisi Esplorativa del dataset Centraline.csv e degli output dei microservizi.

```python
# importazione i moduli base necessari.
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# lettura del dataset dei dati raw delle centraline.
df = pd.read_csv("dati_centraline.csv")
```

```python
# stampa del dataset.
df
```

```python
# stampa statistiche variabili dataset (1)
df.describe()
```

```python
# stampa statistiche variabili dataset (2)
df.info()
```

```python
# visualizzazione valori mancanti per colonna.
df.isnull().sum()
```

```python
# ricerca dati mancanti.
df[df.isnull().any(axis=1)]
```

```python
# eliminazione dati mancanti per riga.
df = df.dropna(axis=0)
```

```python
# verifica eliminazione dati mancanti.
df[df.isnull().any(axis=1)]
```

```python
# ottenimento nomi dei VinVehicle.
VINs = df.VinVehicle.unique()
VINs
```

```python
# eliminazione dei duplicati usando come chiavi la combinazione dei campi VinVehicle, Timestamp e Driver.
df_no_dupl = df.drop_duplicates(subset=["VinVehicle", "Timestamp", "Driver"])
df_no_dupl.shape
```

```python
# visualizzazione dataset preelaborato.
df_no_dupl
```

```python
# rinomino nomi colonne per eliminare caratteri problematici.
df_no_dupl.columns = [i.replace(".", "_") for i in df_no_dupl.columns]
```

```python
# riordino righe sulla base del VinVehicle e Timestamp.
df_no_dupl.sort_values(by=["VinVehicle", "Timestamp"], inplace=True)
```

```python
# visualizzazione dataset pulito e riordinato.
df_no_dupl
```

# ETL2: ANALISI OUTPUT FILEs DA MONGODB COLLECTION SEMPLICE ED ELABORATA

```python
# importazione modulo pymongo per lettura collections semplice ed elaborata da MongoDB.
from pymongo import MongoClient

# Connessione al server di MongoDB.
client = MongoClient('localhost:27017')
db = client["AI4FleetManagement"]
simple_coll = db["Simple"]
elaborated_coll = db["Elaborated"]
```

```python
# Lettura collections derivanti dall'otuput dell'ETL2 (Collection Semplice e Elaborata del DB MongoDB)
# e caricamento con DataFrames Pandas.

df_simple = pd.DataFrame(list(simple_coll.find()))
df_elaborated = pd.DataFrame(list(elaborated_coll.find()))
```

```python
# salvataggio su disco dell'output delle due collections (semplice ed elaborata)
df_simple.to_csv("Output_Simple.csv")
df_elaborated.to_csv("Output_Elaborated.csv")
```

```python
# Preelaborazione dataframes di output per paragone con dataset centraline iniziale.
df_simple.drop("_id", axis=1, inplace=True)
df_simple.drop("MessageDate", axis=1, inplace=True)
df_elaborated.drop("_id", axis=1, inplace=True)
df_elaborated.drop("MessageDate", axis=1, inplace=True)
```

```python
# stampa a video del dataframe della collection semplice.
df_simple
```

```python
# stampa a video del dataframe della collection elaborata.
df_elaborated
```

```python
# Istogrammi di paragone tra dataset originale e coll. semplice relativamente alle variabili
# Odometer e LifeConsumption, per verifica correttezza delle operazioni dei microservizi Injector, ETL1 e ETL2.
#  Gli istogrammi prodotti a partire dai due dataframes sono infatti sovrapponibili.

VINs.sort()
fig, axes = plt.subplots(10, 4 , figsize=(15,20), constrained_layout=True)
for vin in VINs:
    row = int(np.where(VINs == vin)[0][0])

    axes[row, 0].hist(df_no_dupl[df_no_dupl["VinVehicle"]==vin]["Odometer"], 
                        color="red", 
                        label="Odometer")

    axes[row, 1].hist(df_simple[df_simple["VinVehicle"]==vin]["Odometer"], 
                        color="red",
                        label="Odometer")

    axes[row, 2].hist(df_no_dupl[df_no_dupl["VinVehicle"]==vin]["LifeConsumption"],
                        color="blue",
                        label="LifeConsumption")

    axes[row, 3].hist(df_simple[df_simple["VinVehicle"]==vin]["LifeConsumption"], 
                        color="blue",
                        label="LifeConsumption")
    
    axes[row, 0].set(ylabel=vin)

axes[0, 0].set_title("Odometer - starting DF")
axes[0, 1].set_title("Odometer - simple_collection")
axes[0, 2].set_title("LifeConsumption - starting DF")
axes[0, 3].set_title("LifeConsumption - simple_collection")

fig.suptitle("Comparison between starting df and simple collection output histograms", fontsize=16)

plt.savefig("start_vs_simple")
plt.show()
```

```python
# Istogrammi di paragone tra dataset originale, coll. semplice ed elaborata relativamente alla variabile
# DeltaOdometer, per verifica correttezza delle operazioni dei microservizi Injector, ETL1 e ETL2.
# Gli istogrammi prodotti a partire dai tre dataframes sono infatti sovrapponibili.
# I delta dei dataframes originale e della collection semplice sono stati calcolati tramite il metodo .diff() di Pandas.DataFrame,
# mentre quello della collection elaborata è stato calcolato dall'ETL2.

VINs.sort()
fig, axes = plt.subplots(10, 3, figsize=(15,20), constrained_layout=True)
for vin in VINs:
    row = int(np.where(VINs == vin)[0][0])

    axes[row, 0].hist(df_no_dupl[df_no_dupl["VinVehicle"]==vin]["Odometer"].diff(), color="red")

    axes[row, 1].hist(df_simple[df_simple["VinVehicle"]==vin]["Odometer"].diff(), color="blue")

    axes[row, 2].hist(df_elaborated[df_elaborated["VinVehicle"]==vin]["DeltaOdometer"], color="green")
    
    axes[row, 0].set(ylabel=vin)

axes[0, 0].set_title("DeltaOdometer - Starting DF")
axes[0, 1].set_title("DeltaOdometer - Simple_Collection")
axes[0, 2].set_title("DeltaOdometer - Elaborated Collection")

fig.suptitle("Comparison between starting df, simple and elaborated collections outputs histograms\nfor\nDeltaOdometer", fontsize=16)

plt.savefig("start_vs_simple_vs_elab_deltaOdometer")
plt.show()
```

```python
# Istogrammi di paragone tra dataset originale, coll. semplice ed elaborata relativamente alla variabile
# DeltaLifeConsumption, per verifica correttezza delle operazioni dei microservizi Injector, ETL1 e ETL2.
# Gli istogrammi prodotti a partire dai tre dataframes sono infatti sovrapponibili.
# I delta dei dataframes originale e della collection semplice sono stati calcolati tramite il metodo .diff() di Pandas.DataFrame,
# mentre quello della collection elaborata è stato calcolato dall'ETL2.

VINs.sort()
fig, axes = plt.subplots(10, 3, figsize=(15,20), constrained_layout=True)
for vin in VINs:
    row = int(np.where(VINs == vin)[0][0])

    axes[row, 0].hist(df_no_dupl[df_no_dupl["VinVehicle"]==vin]["LifeConsumption"].diff(), color="red")

    axes[row, 1].hist(df_simple[df_simple["VinVehicle"]==vin]["LifeConsumption"].diff(), color="blue")

    axes[row, 2].hist(df_elaborated[df_elaborated["VinVehicle"]==vin]["DeltaLifeConsumption"], color="green")
    
    axes[row, 0].set(ylabel=vin)

axes[0, 0].set_title("DeltaLifeConsumption - Starting DF")
axes[0, 1].set_title("DeltaLifeConsumption - Simple_Collection")
axes[0, 2].set_title("DeltaLifeConsumption - Elaborated Collection")

fig.suptitle("Comparison between starting df, simple and elaborated collections outputs histograms\nfor\nDeltaLifeConsumption", fontsize=16)


plt.savefig("start_vs_simple_vs_elab_deltaLifeConsumption")
plt.show()
```

```python
# rinominazione dataframe iniziale dei dati delle centraline per successive analisi.
df = df_no_dupl
```

```python
# Visualizzazione time series dei Delta Odometer e LifeConsumption di un veicolo esemplificativo (VIN_04) 
# ottenuti dai dataframes originale e delle collection semplici ed elaborata.
# Anche in questo caso i delta dei primi due dataframes sono stati ottenuti con il metodo Pandas.DataFrame.diff()
# mentre per la coll. elaborata questa computazione è stata effettuata dal microservizio ETL2 prima del salvataggio
# su MongoDB.

fig, axes = plt.subplots(2, 3, figsize=(15,8), constrained_layout=True)

axes[0, 0].plot(df[df["VinVehicle"]=="VIN00000000000004"][["Timestamp", "Odometer"]].diff()["Odometer"].values)
axes[0, 0].set_title("df_starting - DeltaOdometer")

axes[0, 1].plot(df_simple[df_simple["VinVehicle"]=="VIN00000000000004"][["Timestamp", "Odometer"]].diff()["Odometer"].values)
axes[0, 1].set_title("df_simple - DeltaOdometer")

axes[0, 2].plot(
    df_elaborated[df_elaborated["VinVehicle"]=="VIN00000000000004"][["Timestamp", "DeltaOdometer"]]["DeltaOdometer"].values)
axes[0, 2].set_title("df_elaborated - DeltaOdometer")

axes[0, 0].set(ylabel="DeltaOdometer")

axes[1, 0].plot(df[df["VinVehicle"]=="VIN00000000000004"][["Timestamp", "LifeConsumption"]].diff()["LifeConsumption"].values)
axes[1, 0].set_title("df_starting - DeltaLifeConsumption")

axes[1, 1].plot(
    df_simple[df_simple["VinVehicle"]=="VIN00000000000004"][["Timestamp", "LifeConsumption"]].diff()["LifeConsumption"].values)
axes[1, 1].set_title("df_simple - DeltaLifeConsumption")

axes[1, 2].plot(
    df_elaborated[df_elaborated["VinVehicle"]=="VIN00000000000004"][["Timestamp", "DeltaLifeConsumption"]]["DeltaLifeConsumption"].values)
axes[1, 2].set_title("df_elaborated - DeltaLifeConsumption")

axes[1, 0].set(ylabel="DeltaLifeConsumption")


fig.suptitle("Comparison between starting df, simple and elaborated collections outputs time series\nfor\nDeltaOdometer and DeltaLifeConsumption in VIN_04.", fontsize=16)


plt.savefig("start_vs_simple_vs_elab_timeSeries")
plt.show()
```
